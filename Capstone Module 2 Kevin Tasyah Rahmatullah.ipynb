## **Background**

**Amazon Web Services (AWS)** adalah cloud yang paling lengkap dan diadopsi secara luas di dunia, menawarkan lebih dari 200 layanan berfitur lengkap dari pusat data secara global. Jutaan pelanggan—termasuk beberapa perusahaan rintisan dengan pertumbuhan tercepat, perusahaan terbesar, dan lembaga pemerintah terkemuka—menggunakan AWS untuk menurunkan biaya, menjadi lebih tangkas, dan berinovasi lebih cepat [(Aws-Amazon, 2024)](https://aws.amazon.com/id/what-is-aws/).

Tahun-tahun awal AWS ditandai dengan inovasi dan pertumbuhan yang pesat. Pada tahun 2007, AWS meluncurkan layanan besar pertamanya, Amazon S3 (Simple Storage Service), yang menyediakan solusi penyimpanan yang skalabel dan tahan lama untuk bisnis. Hal ini diikuti dengan peluncuran Amazon EC2 (Elastic Compute Cloud) pada tahun 2008, yang menyediakan sumber daya komputasi terukur di cloud [(Great Learning, 2024)](https://www.mygreatlearning.com/aws/tutorials/history-of-aws). Ini mencakup gabungan antara infrastructure-as-a-service (IaaS), platform-as-a-service (PaaS) dan paket software-as-a-service (SaaS). AWS menawarkan alat seperti daya komputasi, penyimpanan database, dan layanan pengiriman konten [(TechTarget, 2024)](https://www.techtarget.com/searchaws/definition/Amazon-Web-Services).

Layanan infrastruktur cloud yang dikenal sebagai **Infrastructure as a Service (IaaS)** terdiri dari sumber daya komputasi yang sangat scalable dan terotomatisasi. IaaS pada dasarnya bersifat self-service untuk mengakses dan memantau komputer, jaringan, storage, dan layanan lainnya. Layanan platform cloud, yang juga dikenal sebagai **Platform as a Service (PaaS)**, menyediakan komponen cloud untuk perangkat lunak tertentu saat digunakan terutama untuk aplikasi. PaaS memberikan framework untuk pengembang yang dapat dibangun dan digunakan untuk membuat aplikasi yang disesuaikan dengan kebutuhan masing-masing pengguna. **Software as a Service (SaaS)**, atau dikenal juga sebagai layanan aplikasi cloud, mewakili opsi yang paling umum digunakan untuk bisnis. SaaS menggunakan sarana internet untuk mengirimkan aplikasi yang dikelola oleh vendor pihak ketiga kepada penggunanya. Sebagian besar aplikasi SaaS berjalan langsung melalui browser web Anda, yang berarti Anda sebagai pengguna tidak memerlukan pengunduhan atau penginstalan apa pun [(CBNCloud, 2024)](https://cbncloud.co.id/about/publication-detail/iaas-saas-paas-pengertian-perbedaan-keunggulan-keuntungan-masing-masing-model-layanan-cloud-untuk-penerapan-dalam-lingkup-bisnis).

### **3. Outliers**

Outliers adalah observasi yang terletak pada jarak abnormal dari nilai lain dalam sampel acak dari suatu populasi. Dalam arti tertentu, definisi ini diserahkan kepada analisis (atau proses konsensus) untuk memutuskan apa yang dianggap abnormal. Sebelum observasi abnormal dapat diidentifikasi, observasi normal perlu dikarakterisasi [(NIST)](https://www.itl.nist.gov/div898/handbook/prc/section1/prc16.htm). Outlier dapat terjadi karena disinformasi dari subjek, error respon subjek, atau kesalahan entri data. Dalam beberapa kasus, jelas bahwa outlier harus dihapuskan karena bisa menimbulkan kesalahan analisis, namun di kasus yang lain, hal ini mungkin disebabkan oleh standar atau keputusan penilaian di mana outlier merupakan penyimpangan alami. Sebagai Data Analyst kita perlu mencari tahu outliers karena satu outlier saja dapat secara dramatis mempengaruhi rata-rata dan mengganggu hasil akhir analisis [(TechTarget, 2018)](https://www.techtarget.com/whatis/definition/outlier).

**Mencari Outliers dengan Menggunakan Boxplot**

Box plot, juga dikenal sebagai plot kotak, atau plot kotak-dan-kumis, adalah cara standar untuk menampilkan distribusi dataset berdasarkan ringkasan titik data lima angka: “minimum”, kuartil pertama [Q1], median, kuartil ketiga [Q3], dan “maksimum”. Boxplot juga merupakan grafik yang memberi tahu suatu perusahaan bagaimana nilai dataset perusahaan tersebut tersebar. Box plot dapat memberi tahu perusahaan tentang outlier dan berapa nilainya. Box plot juga dapat memberi tahu perusahaan apakah dataset tersebut simetris, seberapa erat pengelompokan dataset tersebut, dan bagaimana dataset tersebut menyimpang (atau skewed) [(BuiltIn, 2023)](https://builtin.com/data-science/boxplot).

```python
outliers_info = {} # Memakai dict untuk memasukkan informasi outliers

for column in ['Sales', 'Quantity', 'Discount', 'Profit']:
    # Kalkulasikan Q1, Q3, dan IQR
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    # Kalkulasikan ambang batas (bounds)
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Identifikasikan outliers
    outliers_condition = (df[column] < lower_bound) | (df[column] > upper_bound)
    outliers = df[outliers_condition]

    # Menyimpan informasi di Dict
    outliers_info[column] = {
        'lower_bound': lower_bound,
        'upper_bound': upper_bound,
        'number_of_outliers': outliers_condition.sum(),  # Jumlah total outliers
        'outliers_indices': outliers.index.tolist()  # Indices of outliers
    }

# Ringkasan ke outliers_summary

outliers_summary = {key: {'bounds': (value['lower_bound'], value['upper_bound']), 
                          'number_of_outliers': value['number_of_outliers']} 
                    for key, value in outliers_info.items()}

# Membuat DataFrame dari 'outliers_summary'
outliers_summary_df = pd.DataFrame.from_dict(outliers_summary, orient='index')

# Membagi 'bounds' menjadi 'lower_bound' dan 'upper_bound' di kolom
outliers_summary_df[['lower_bound', 'upper_bound']] = outliers_summary_df['bounds'].apply(pd.Series)

# Menghapus kolom "bounds"
outliers_summary_df.drop('bounds', axis=1, inplace=True)

# Menyusun kembali kolom-kolom
outliers_summary_df = outliers_summary_df[['lower_bound', 'upper_bound', 'number_of_outliers']]

# Menampilkan DataFrame ke styled_table
styled_table = outliers_summary_df.style.format('{:.1f}')
styled_table
```

### **5. Uji Normalitas Data**

Distribusi normal, disebut juga distribusi Gaussian, adalah distribusi probabilitas yang simetris terhadap mean, yang menunjukkan bahwa data yang dekat dengan mean lebih sering muncul dibandingkan data yang jauh dari mean [(Investopedia, 2024)](https://www.investopedia.com/terms/n/normaldistribution.asp).

**1) Uji Normalitas menggunakan func. normaltest**

```python
df_numerikal = df[['Sales','Quantity', 'Profit', 'Discount']] # Membuat variabel baru berisi kolom Sales, Quantity, Profit, dan Discount
for setiap_kolom in df_numerikal:
    stat, pval = normaltest(df[setiap_kolom]) # Uji normalitas menggunakan normaltest
    if pval < 0.05:
        print(f'Nilai pval = {pval} dan nilai statistik = {stat}')
        print(f'Di kolom {setiap_kolom}, tidak ada cukup bukti untuk membuktikan bahwa distribusinya normal')
        print('DISTRIBUSI TIDAK NORMAL') # Ha normalitas
        print('\n')
    else:
        print(f'Nilai p-val = {pval} dan nilai statistik = {stat}')
        print(f'Di kolom {setiap_kolom}, terdapat cukup bukti untuk membuktikan bahwa distribusinya normal')
        print('DISTRIBUSI NORMAL') # H0 normalitas
        print('\n')
```

Dari hasil normaltest tersebut, bisa disimpulkan bahwa kolom kolom Sales, Quantity, Profit, dan Discount tidak terdistribusi normal dikarenakan semua nilai P-Valuenya kurang dari 0.05. Maka distribusi data AWS SaaS bersifat **distribusi tidak normal**. Untuk meyakinkan lagi, kita harus membuat visualisasinya.

**2). Uji Normalitas menggunakan Visualisasi Histogram**

Menganalisis tren Sales membantu memahami kinerja produk, layanan, atau bisnis dari waktu ke waktu. Kita menggunakan **median** karena dataset AWS SaaS berdistribusi **tidak normal**.

```python
import requests
from geopy.geocoders import Nominatim

def get_country_coordinates(country_name):
    geolocator = Nominatim(user_agent="geoapi")
    
    # Gunakan request tanpa verifikasi SSL
    session = requests.Session()
    # session.verify = False
    
    url = f"https://nominatim.openstreetmap.org/search?q={country_name}&format=json&limit=1"
    response = session.get(url)
    if response.ok:
        data = response.json()
        if data:
            latitude = float(data[0]['lat'])
            longitude = float(data[0]['lon'])
            return latitude, longitude
    return None, None

# Anggaplah dfri_country_storage adalah hasil dari operasi sebelumnya
# Tambahkan kolom latitude dan longitude ke dfri_country_storage
latitude = []
longitude = []
for i in dfri_country_storage['Country']:
    lat, long = get_country_coordinates(i)
    latitude.append(lat)
    longitude.append(long)

dfri_country_storage['Latitude'] = latitude # Masukkan hasil latitude ke kolom dfri_country_storage
dfri_country_storage['Longitude'] = longitude # Masukkan hasil longitude ke kolom dfri_country_storage

dfri_country_storage
```

#### **Uji Hipotesis antara Segmen Customer dengan Sales**

```python
smb_sales = df_storage[df_storage['Segment'] == 'SMB']['Sales']
strategic_sales = df_storage[df_storage['Segment'] == 'Strategic']['Sales']
enterprise_sales = df_storage[df_storage['Segment'] == 'Enterprise']['Sales']

# Melakukan tes Kruskal-Wallis karena distribusi datanya tidak normal
H0 = "Tidak ada perbedaan antara nilai Sales dengan setiap segmen customer"
Ha = "Ada salah satu segmen customer yang mempunyai perbedaan nilai Sales"
Alpha = 0.05
h_stat, p_value = kruskal(smb_sales, strategic_sales, enterprise_sales)

print(f"Kruskal-Wallis Test:\nH-statistic: {h_stat}\nP-value: {p_value}")
if p_value < Alpha:
    print("Menolak H0")
    print(Ha)
else:
    print("Menerima H0")
    print(H0)
```

**Insight**

Berdasarkan dari hasil uji hipotesis, hasilnya menunjukkan bahwa **tidak ada perbedaan yang signifikan** antara median Sales terhadap segmentasi customer (SMB, Strategic, Enterprise).

**P-Value yang didapatkan sebesar 0,791, yang mana nilai tersebut lebih tinggi daripada level signifikansinya (Alpha) sebesar 0.05. Artinya adalah hasil ini tidak ada cukup bukti untuk menolak H0, yang menyatakan bahwa tidak ada perbedaan yang signifikan antara median Sales terhadap segmentasi customer (SMB, Strategic, Enterprise).**

Hasil uji Kruskal-Wallis menunjukkan bahwa, dengan mempertimbangkan segmen (SMB, StrategiC, dan Enterprise) dengan nilai Sales, tidak ada perbedaan median Sales yang signifikan di antara segmen-segmen tersebut.

Artinya, nilai median Sales dengan ketiga segmen tersebut (SMB, Strategic, Enterprise) tidak berbeda secara signifikan satu sama lain. Oleh karena itu, dalam konteks nilai Sales, nampaknya semua segmen ini mungkin menghadapi situasi serupa mengenai tingkat median Salesnya, tanpa ada satu segmen yang secara signifikan berkinerja buruk atau mengungguli segmen lainnya dalam hal Sales.

### **Korelasi Analisis**

```python
# Membuat kode korelasi
numerical_col = ["Sales", "Discount", "Profit", "Sales Margin",'Quantity'] # Kolom-kolom yang ingin dicari nilai korelasi
correlation = df_storage[numerical_col].corr(method = "spearman")
hide = np.triu(correlation) # Untuk menyembunyikan nilai korelasi yang terlihat double (jadi terlihat unique value dari nilai korelasi)
display(correlation)

# Membuat Heatmap
plt.figure(figsize = (12, 6))
sns.heatmap(
    data = correlation,
    annot = True,
    fmt = ".2f",
    cmap = "magma",
    center = 0,
    mask= hide
)

plt.title("Korelasi antara Sales vs Discount vs Sales Margin vs. Profit di produk Storage", size=16, pad=10, fontweight='bold') # Judul Heatmap
plt.show()
```
